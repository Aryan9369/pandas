```python
import numpy as np
import pandas as pd
```

```python
delivery=pd.read_csv("deliveries.csv")
delivery
```

```python
match=pd.read_csv('matches.csv')
match
```

```python
#we use merge function when we have to need merge two data set
#like here we want only orange cap player of evry season which is not possible using only one dataset that's why need to merge this dataset
#we are giving the column name on which we are going to merge dataset 
new_match=delivery.merge(match,left_on='match_id',right_on='id') #this is inner merge mean only when column is matched only that row is picked otherwise it will dropped
```

```python
new_match #it has all row mean no data loss here #but column in increased
```

```python
#now we  have to make goup on the basis of season and player bcs we need highest run's player of each season so, first we have to find player wise each season
x=new_match.groupby(['season','batsman']) #heer we have two index column first is season and second is batsman
```

```python
#now we will add the run of each of of each player
#we want only seaon and batsman who has orage cap in the season
#it also return an dataframe ,lets use it with sum of some column of dataframe on the basis of player in a season
x.sum()
```

```python
#now we add runs of each column and it will return a series bcs we are fetching column batsman_runs
#we also sort it ont the basis of batsman_run so that we can see who is on the top in each season 
x.sum()['batsman_runs'].sort_values(ascending=False)
```

```python
#now we will convert it into dataframe to do that we use a funtion called reset_index()
y=x.sum()['batsman_runs'].sort_values(ascending=False).reset_index()
y
```

```python
#now we will drop all the players of each season except first occurred player of each season bcs we already sort it
#so we are going to use drop_duplicate(subset='season',keep='first') #1 keep only first #subset mean on which basis it will see duplicate
z=y.drop_duplicates('season').sort_values('season',ascending=False)
z#we also sorting it on the basis of season that we get a good dataframe
```
