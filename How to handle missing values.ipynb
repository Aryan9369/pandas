```python
#we have two method to do it 
#dropna()-drop the missing values
#fillna()-fillthe missing values
```

```python
import numpy as np 
import pandas as pd
```

```python
data=pd.read_csv('Titanic-Dataset.csv')
data
```

```python
#we can see the shape before cleaning the data
data.shape
```

```python
#no we are going to use dropna()
data.dropna()
```

```python
#shape after the cleaning
data.dropna().shape #means only 183 rows have no null value in any column
```

```python
#if we want to apply it column wise we can use it 
data.dropna(axis=1).shape#bydefault it's 0
#meeans 9 columns in which no  values is null in any of it's row
```

```python
#it have some parameter also like how='any' by default means if any one column or row have null then it will detele whole row or column
#if we use how='all' the if all the values are null then it will delete that column or row
data.dropna(axis=0,how='all')
```

```python
data.dropna(axis=0,how='all').shape  #you can see the shape means no row  has all null values
```

```python
data.dropna(axis=1,how='all').shape #same for column
```

```python
#there s another parameter subset if you want to apply cleaning on some column not all data
#pass the list of column
data.dropna(axis=0,how='any',subset=["Cabin","Embarked"]) #if anyone of them has null that row will drop
```

```python
#there is another thresh you can apply threshold means minimum number of non-NULL values or NaN should be in a row or collumn
data.dropna(thresh=11)
```

```python
#you can see shape in last that 733 rows only have at least 11 non-NULL which droped they have less that 11 non-NUll values or we can say more than one NULL value
data.dropna(thresh=11).shape
```

```python
#NOW
```

```python
#NOW WE ARE GOING TO SEE THE FILLNA() METHOD
data.fillna(0) #means every NaN values will be replaced by 0 we can different valuee also
#but it is not appropriate to use directly on whole dataframe instead of it , we should use it on perticular column bcs string datadype column should not contain 0
```

```python
#so we are going to apply it on perticular column
#apply on cabin column bcs its string type 
data['Cabin'].fillna('NOT MENTIONED',inplace=True)#it will return a series you can make it permanent changes
```

```python
#now you can see it in dataframe
data
```

```python
#there is also soome parameter like axis ot method
#lets see method 
#it will take two values bfill or ffill #IN bfill it replace the null value with its value after it
#but in ffill it replce the value which is before it

# OLD way (will give FutureWarning)
# data.fillna(method='bfill')

# NEW way (correct)
data=data.bfill()
```
