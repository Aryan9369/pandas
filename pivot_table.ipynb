```python
import numpy as np
import pandas as pd
```

```python
#we are going to work on pivot table
#it help you  to see data from diffrent angle and to organise and summarize it
 
delivary=pd.read_csv('deliveries.csv')
delivary.head()
```

```python
#now we are going to apply pivot table
#we want to see six on each over per team
#first we remove unecessary data from our data 
mask=delivary['batsman_runs']==6
six=delivary[mask]
six#now you can see data  with only six
```

```python
pt=six.pivot_table(index='batting_team',columns='over',values='batsman_runs',aggfunc='count')
pt#you can change row and column for this data
```

```python
#now we are going to apply heat map on it
#first import seaborn library
import seaborn as sns
```

```python
#seaborn has a function called heatmap which takes only a pivot table
sns.heatmap(pt) #you can see the a scale also that shows the intensity of colour according to number of sixes
```
