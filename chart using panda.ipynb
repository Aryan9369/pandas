```python
#value_counts is a function that can used in a series to count numbers of same category
import pandas as pd 
import numpy as np
```

```python
data=pd.read_csv("matches.csv")
data
```

```python
#lets fetch one column to get a series #left is index
data["winner"]
```

```python
#lets how many times each team has been win
data["winner"].value_counts() #left is index which is unique and right is value
```

```python
#we can draw different types of chart using the plot function in pandas
import matplotlib.pyplot as plt #important
 
data["winner"].value_counts().plot(kind="bar")
```

```python
#you can print it as horizontal
data["winner"].value_counts().plot(kind="barh") #use barh
```

```python
#for pie chart of only first five
data["winner"].head().value_counts().plot(kind="pie")
```

```python
#histogram is you have to draw a interval based chart #we can aply it on numerical series not necessary for catergorical series
data["win_by_runs"].plot(kind="hist")#no need to use value_counts because here we have no catergories to make categorical series
```

```python
#we can also use bar graph directly but result will be disturbing 😂
data["win_by_runs"].plot(kind="bar")
```
